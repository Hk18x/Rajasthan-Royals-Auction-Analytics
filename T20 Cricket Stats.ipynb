```python
import pandas as pd
import numpy as np
```

```python
t20_2023 = pd.read_csv('t20_2023.csv')
t20_2024 = pd.read_csv('t20_2024.csv')
t20_2022 = pd.read_csv('t20_2022.csv')
```

```python
t20_merged = pd.concat([t20_2022, t20_2023, t20_2024],ignore_index = True)
```

```python
# Define the teams of interest
teams = ['india', 'australia', 'england', 'south africa', 'srilanka', 'bangladesh', 'afghanistan', 'west indies']

# Filter the data
openers_powerplay_data = t20_merged[
    (t20_merged['team_bat'].str.lower().isin(teams)) &  # Only selected teams
    (t20_merged['over'] >= 1) & (t20_merged['over'] <= 6)
]

openers_powerplay_data
```

```python
powerplay_top_scorers = openers_powerplay_data.groupby('bat').agg(runs=('batruns','sum'), balls_taken=('ballfaced','sum'),
                                                           dismissals= ('dismissal','count')).sort_values(by='runs', ascending = False).reset_index()
powerplay_top_scorers['Average'] = round(powerplay_top_scorers['runs']/powerplay_top_scorers['dismissals'],2)
powerplay_top_scorers['Strike Rate'] = round((powerplay_top_scorers['runs']/powerplay_top_scorers['balls_taken'])*100,2)

powerplay_top_scorers.sort_values(by = 'Strike Rate', ascending = False)
```

```python
# Filter for players with runs > 100 and Strike Rate > 160
filtered_top_scorers = powerplay_top_scorers[
    (powerplay_top_scorers['runs'] > 100) & 
    (powerplay_top_scorers['Strike Rate'] > 160)
]

# Sort the filtered data by 'Strike Rate' in descending order
filtered_top_scorers = filtered_top_scorers.sort_values(by='Strike Rate', ascending=False)
filtered_top_scorers
```
