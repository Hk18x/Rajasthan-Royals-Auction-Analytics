### WIN IMPACT

```python
batsman_match_runs = rr_df_bat.groupby(['p_match', 'bat'])['score'].sum().reset_index()
batsman_match_runs.rename(columns={'score': 'batsman_runs'}, inplace=True)
```

```python
team_match_runs = rr_df_bat.groupby('p_match')['score'].sum().reset_index()
team_match_runs.rename(columns={'score': 'team_runs'}, inplace=True)
```

```python
batsman_match_contribution = batsman_match_runs.merge(team_match_runs, on='p_match', how='left')
```

```python
batsman_match_contribution.head()
```

```python
batsman_match_contribution['contribution_percentage'] = (batsman_match_contribution['batsman_runs'] / batsman_match_contribution['team_runs']) * 100
```

```python
match_results = {
    1422122: 'won', 1422127: 'won', 1422132: 'won', 1422137: 'won',
    1426262: 'lost', 1426265: 'won', 1426269: 'won', 1426276: 'won',
    1426282: 'won', 1426288: 'lost', 1426294: 'lost', 1426299: 'lost',
    1426303: 'lost', 1426310: 'won', 1426311: 'lost'
}
rr_df_bat['Result'] = rr_df_bat['p_match'].map(match_results)
winning_matches = rr_df_bat[rr_df_bat['Result'] == 'won']
winning_match_ids = winning_matches['p_match'].unique()
winning_contribution = batsman_match_contribution[batsman_match_contribution['p_match'].isin(winning_match_ids)]
```

```python
average_win_contribution = winning_contribution.groupby('bat')['contribution_percentage'].mean().reset_index()
average_win_contribution.rename(columns={'contribution_percentage': 'win_impact_percentage'}, inplace=True)
```

```python
reference_sr = 120
winning_matches = rr_df_bat[rr_df_bat['Result'] == 'won']
winning_stats = winning_matches.groupby('bat').agg(
    winning_runs=('batruns', 'sum'),
    winning_balls=('ballfaced', 'sum')
).reset_index()
winning_stats['winning_strike_rate'] = (winning_stats['winning_runs'] / winning_stats['winning_balls']) * 100
winning_stats['weighted_runs'] = winning_stats['winning_runs'] * (winning_stats['winning_strike_rate'] / reference_sr)
total_runs = rr_df_bat.groupby('bat')['score'].sum().reset_index().rename(columns={'score': 'total_runs'})
batsman_impact = pd.merge(winning_stats, total_runs, on='bat', how='left')
batsman_impact['win_impact'] = (batsman_impact['weighted_runs'] / batsman_impact['total_runs']) * 100
```

```python
total_win_impact = batsman_impact['win_impact'].sum()
batsman_impact['normalized_win_impact'] = (batsman_impact['win_impact'] / total_win_impact) * 100
batsman_impact[['bat', 'total_runs', 'winning_runs', 'winning_strike_rate', 'weighted_runs', 'win_impact', 'normalized_win_impact']].sort_values(by='normalized_win_impact', ascending=False)
```
